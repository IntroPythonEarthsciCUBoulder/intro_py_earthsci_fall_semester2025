There is an ongoing debate about how river bank erosion may be affected in rivers in permafrost landscapes. *Permafrost* is defined as soil that remains frozen through two consequetive summers, except for its very top layer that seasonally thaws (the active layer). This means that in permafrost landscapes, at some depth all the sediment or substrate is frozen, for example in Northern Alaska this can be as little as 40 cm. <br>

How does a river carve into frozen banks in permafrost landscapes? Do bank even get eroded? It turns out the river partly melts the bank material, while in full summer it may obey more normal sediment transport conditions at least for that top layer.
    
Satellite data analysis allows mapping of river banks over time, and such analysis appears to show slower rates of bank erosion in Arctic permafrost terrain as compared to other rivers in warmer regions worldwide (Rowland et al., 2023). We here look at a data plot from the satelite data analysis study on bank erosion in permafrost rivers in contrast to rivers in lower latitudes (temperate and tropical regions) (Rowland et al., 2023). You can find this paper referenced down below.




It plots river erosion and river streampower for river in non-permafrost terrain (circles), and in permafrost terrain (squares). The colors indicate climate zones, for now, that we are not using that information.

The study found that the relationship for bank erosion (E) for normal rivers, as a function of local or reach streampower (the grey line in the plot) is a power function: 

$E  =  0.008  \Omega ^{0.77} $ 

$ \Omega$ in $W/m^{-1}$ <br>
$ E $ in $ m/y $


Then, the study found that the relationship for bank erosion for permafrost rivers, as a function of local or reach streampower (the blue line in the plot) again is a power function: 

$E  =  0.19  \Omega ^{0.23} $

You can see how these permafrost rivers (squares) deviate from the larger dataset (circles) and have lower values of erosion for the same streampowers, especially in stretches with large streampower.
